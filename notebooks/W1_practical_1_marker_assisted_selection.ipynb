## Some background before we start

Plant breeders want to combine desirable traits (e.g. disease resistance, drought tolerance, or high yield) into a single crop variety. Traditionally this is done by crossing plants and simply looking at which offspring *look* the best (this is called phenotypic selection). That is slow: some traits only show up late in a plant's life, and some are expensive or destructive to measure (e.g. you'd have to expose a plant to salt stress just to see if it survives).

**Genetic markers** offer a shortcut. A marker is a specific, measurable position in a plant's DNA where individuals can differ from one another. In this notebook we simplify this to just two possible values, `0` and `1`, for every marker:

- `0` = this plant does **not** carry a particular DNA variant at that position
- `1` = this plant **does** carry that DNA variant

If we know that a marker is reliably linked to a trait we care about, we can screen very young plants (often from just a small leaf sample, days after germination) with a lab test (genotyping), instead of waiting months to observe the fully grown plant. This is the core idea behind marker assisted selection: use DNA markers to predict which plants are worth keeping, long before you could otherwise tell.

In this notebook we will:
1. Use a statistical method, a genome-wide association study (GWAS), to find a genetic marker associated with salt resistance
2. Use that marker to cross a high-yielding but salt-sensitive crop with a salt-resistant one, making it more salt resistant but losing minimal yield

## Step 2: Find markers associated with the trait (a GWAS)

To find out which marker(s) are responsible for salt resistance, we run a **Genome-Wide Association Study (GWAS)**. In short, a GWAS statistically tests, separately for *each* marker, whether plants with a `1` at that marker tend to have a different phenotype than plants with a `0`. The result of this test is a p-value: a small p-value means it is unlikely that the difference we observe between the two groups happened purely by chance, so it's more likely to reflect a real, biological effect.

Because p-values can become extremely small, we usually plot `-log10(p-value)` instead of the p-value itself: this turns tiny p-values into large, easy-to-compare bars. The higher the bar, the stronger the evidence that a marker is associated with the trait. A plot of this across all markers is called a Manhattan plot, because the tall bars sticking up resemble a city skyline.

```python
my_population.show_manhattan_plot(to_phenotype='Salt Resistance (% survival)')
```

## Step 3: Introgress the marker into an elite variety

Now that we've pinpointed a marker for salt resistance, we want to move that specific piece of DNA into a plant that farmers already grow: one with a high yield, but which lacks the marker. This process is called **introgression**: transferring a gene (or a small region of a chromosome) from one variety into another through repeated crossing and selection.

We'll use two starting populations:

- a **resilient population**: genetically diverse and carrying the salt-resistance marker. Think of it as a wild relative or old landrace: hardy, but usually a poor yielder.
- an **agricultural population**: today's elite, high-yielding variety that farmers actually plant, but which lacks the salt-resistance marker.

```python
resilient_population = get_resilient_population()
resilient_population.show_size()
# Now check if it indeed contains the marker at your location
resilient_population.show_marker_at_location(FILL_IN_INDEX)
```

```python
agricultural_population = get_agricultural_population()
agricultural_population.show_marker_at_location(FILL_IN_INDEX)
```

Let's check how the two starting plants differ in phenotype:

```python
# check difference in phenotypes
resilient_population.show_all_phenotypes()
agricultural_population.show_all_phenotypes()
```

Notice the trade-off: one plant is salt resistant but low-yielding, the other is high-yielding but salt sensitive. That trade-off is exactly what we're about to break, by combining the best of both.

## Step 4: Marker assisted backcrossing

We now perform marker assisted backcrossing: repeatedly crossing back to the high-yielding agricultural variety, while using the marker to always keep the salt-resistance gene from the resilient donor. Each round:

1. Cross the current population with the recurrent parent (the agricultural population)
2. Genotype the offspring, and keep *only* the plants that carry the desired marker allele
3. Repeat

Over successive rounds, this recovers more and more of the recurrent parent's genome (and therefore its high yield), while keeping the one gene we care about.

```python
new_population = perform_cross_between(resilient_population, agricultural_population, n_offspring=10)
selected_population = new_population.select_plants_with_marker_at_location(FILL_IN_INDEX, desired_allele=1)
selected_population.n_plants
```
